```julia
S = AbstractFBCModels.stoichiometry(model)
```

```julia
# Get the stoichiometry matrix
n_reactions = size(S, 2)

# Store all unique complexes
all_complexes = Set()

for rxn_idx in 1:n_reactions
    rxn_col = S[:, rxn_idx]
    
    # Find substrate metabolites (negative stoichiometric coefficients)
    substrate_mets = findall(x -> x < 0, rxn_col)
    if !isempty(substrate_mets)
        substrate_stoich = -rxn_col[substrate_mets]
        push!(all_complexes, (substrate_mets, substrate_stoich))
    end
    
    # Find product metabolites (positive stoichiometric coefficients)
    product_mets = findall(x -> x > 0, rxn_col)
    if !isempty(product_mets)
        product_stoich = rxn_col[product_mets]
        push!(all_complexes, (product_mets, product_stoich))
    end
end

# Number of unique complexes
length(all_complexes)
println(all_complexes)
```

```julia
# Check if we have complexes that are the same but in different orders
# Convert to sorted tuples for comparison
function normalize_complex(mets, stoich)
    # Sort by metabolite index
    perm = sortperm(mets)
    return (mets[perm], stoich[perm])
end

# Rebuild with normalized complexes
normalized_complexes = Set()
for rxn_idx in 1:n_reactions
    rxn_col = S[:, rxn_idx]
    
    substrate_mets = findall(x -> x < 0, rxn_col)
    if !isempty(substrate_mets)
        substrate_stoich = -rxn_col[substrate_mets]
        push!(normalized_complexes, normalize_complex(substrate_mets, substrate_stoich))
    end
    
    product_mets = findall(x -> x > 0, rxn_col)
    if !isempty(product_mets)
        product_stoich = rxn_col[product_mets]
        push!(normalized_complexes, normalize_complex(product_mets, product_stoich))
    end
end

println("Original count: ", length(all_complexes))
println("Normalized count: ", length(normalized_complexes))
println("Difference: ", length(all_complexes) - length(normalized_complexes))
```

```julia
# Check if the model is already irreversible
rxn_ids = AbstractFBCModels.reactions(model)
reversible_count = 0
all_bounds = AbstractFBCModels.bounds(model)
lbs, ubs = all_bounds
for (i, rxn_id) in enumerate(rxn_ids)
    lb = lbs[i]
    ub = ubs[i]
    if lb < 0 && ub > 0
        reversible_count += 1
    end
end
println("Total reactions: ", length(rxn_ids))
println("Reversible reactions: ", reversible_count)
println("Irreversible reactions: ", length(rxn_ids) - reversible_count)
```

```julia
# Extract matrices and save using MatrixMarket format
using SparseArrays, DelimitedFiles

println("=== Extracting Matrices ===")
println()

# Extract Y, A, and S matrices
println("Building Y (complex stoichiometry)...")
Y, metabolite_ids, complex_ids = complex_stoichiometry(model, return_ids=true)
println("  Y dimensions: ", size(Y))
println("  Y non-zeros: ", nnz(Y))
println()

println("Building A (incidence matrix)...")
A, complex_ids_A, reaction_ids = incidence(model, return_ids=true)
println("  A dimensions: ", size(A))
println("  A non-zeros: ", nnz(A))
println()

println("Getting S (stoichiometry matrix)...")
S = AbstractFBCModels.stoichiometry(model)
println("  S dimensions: ", size(S))
println("  S non-zeros: ", nnz(S))
println()

# Verify Y*A = S
println("Verifying Y*A = S...")
S_computed = Y * A
println("  Y*A == S: ", S_computed == S)
println()

# Save directory
save_dir = "/work/schaffran1/results_testjobs/for_babak"

# Save matrices using MatrixMarket
using MatrixMarket
println("Saving matrices to MatrixMarket format...")
MatrixMarket.mmwrite(joinpath(save_dir, "Y_metabolite_complex.mtx"), Y)
println("  ✓ Saved Y_metabolite_complex.mtx")

MatrixMarket.mmwrite(joinpath(save_dir, "A_complex_reaction.mtx"), A)
println("  ✓ Saved A_complex_reaction.mtx")

MatrixMarket.mmwrite(joinpath(save_dir, "S_stoichiometry.mtx"), S)
println("  ✓ Saved S_stoichiometry.mtx")

# Save ID mappings as CSV
println()
println("Saving ID mappings...")
writedlm(joinpath(save_dir, "metabolite_idx.csv"), hcat(1:length(metabolite_ids), String.(metabolite_ids)), ',')
println("  ✓ Saved metabolite_idx.csv")

writedlm(joinpath(save_dir, "complex_idx.csv"), hcat(1:length(complex_ids), String.(complex_ids)), ',')
println("  ✓ Saved complex_idx.csv")

writedlm(joinpath(save_dir, "reaction_idx.csv"), hcat(1:length(reaction_ids), String.(reaction_ids)), ',')
println("  ✓ Saved reaction_idx.csv")

println()
println("=== Summary ===")
println("Number of metabolites: ", length(metabolite_ids))
println("Number of complexes: ", length(complex_ids))
println("Number of reactions: ", length(reaction_ids))
println()
println("All matrices saved to: $save_dir")
```
